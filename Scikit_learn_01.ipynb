# Scikit Learn Library Basic Usage


```python
# libraries import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# simple linear regrassion

from sklearn.linear_model import LinearRegression
```

```python
# Load Dataset
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x='total_bill' , y='tip', data=df)
```

```python
# create a regression line plot using seaborn

sns.lmplot(x='total_bill', y='tip',data=df)
```

```python
# Now Choose X and Y
# For Linear regrassion must be Numaric X and Y
# we pridict tip based on total_bill , than we say totall_bill they pridict tip
# Now
# Select the Feature Column from df

X = df[['total_bill']]
# select the target column form df
y = df['tip']
```

```python
# Load the Linear Model regression model

model = LinearRegression()
#fit the model means train the model
model.fit(X,y)
```

```python
# lets predict the tip for an unknown total_bill

model.predict([[20]])
```
